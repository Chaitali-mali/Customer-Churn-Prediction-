```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
churn_df = pd.read_csv(r"C:\Users\chait\Downloads\Churn_Modelling.csv")
```

```python
print(churn_df.head())
print(churn_df['Exited'].value_counts())
print(churn_df['Exited'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Exited', data=churn_df, palette='Set2')
plt.title('Churn (Exited) Distribution')
plt.xlabel('Exited')
plt.ylabel('Count')
plt.show()
```
